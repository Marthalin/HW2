Visit https://www.indeed.com/Top-Rated-Workplaces, and search for Data Scientist in Los Angeles. Scrape 50 company reviews of the first 50 companies (if fewer than 50, just get the companies names on the first page). Retrieve the company’s name, title of the reviewer, the date of the review, the title, rating, review content, pros, and cons and the review. The output can to be shown in the form of a table. If any of the companies doesn’t have 50 reviews, then you just need to get whatever it has on the website.

```python
from bs4 import BeautifulSoup
import urllib
import re
import pandas as pd
import requests
from urllib.request import urlopen
```

```python
company = []
reviewer = []
date = []
title = []
rating = []
content = []
pros = []
cons = []
url = "https://www.indeed.com/cmp?q=Data+Scientist&l=Los+Angeles%2C+CA&from=discovery-cmp-search"

page = urlopen(url)
soup = BeautifulSoup(page, 'lxml')
all_matches = soup.findAll(attrs = {'data-tn-element':['reviews-footer-link']})

for review in all_matches:
    for i in range(3):
        review_url = "https://www.indeed.com" + review['href']
        review_page = urlopen(review_url)
        review_soup = BeautifulSoup(review_page, 'lxml')
        review_container = review_soup.findAll(attrs = {'class':['cmp-review-container']})

        for each in review_container:
            #company name
            company_name = review_soup.findAll(attrs = {'class':['cmp-company-name']})
            company.append(company_name[0].text)
            #review title
            review_title = each.findAll('div', attrs = {'class':['cmp-review-title']})
            title.append(review_title[0].text)
            #rating
            review_rating = each.findAll(attrs = {'class':['cmp-ratingNumber']})
            rating.append(review_rating[0].text)
            #review content
            review_content = each.findAll(attrs = {'class':['cmp-review-text']})
            content.append(review_content[0].text)
            #date
            review_date = each.findAll(attrs = {'class':['cmp-review-date-created']})
            date.append(review_date[0].text)
            #title of reviewer
            review_reviewer = each.findAll(attrs = {'class':['cmp-reviewer']}) 
            reviewer.append(review_reviewer[0].text)

            #Pros
            if not each.findAll(attrs = {'class':['cmp-review-pros']}):
                pros.append('N/A')
            else:
                review_pros = each.findAll(attrs = {'class':['cmp-review-pros']})
                pros.append(review_pros[0].text)

            #Cons
            if not each.findAll(attrs = {'class':['cmp-review-cons']}):
                cons.append('N/A')
            else:
                review_cons = each.findAll(attrs = {'class':['cmp-review-cons']})
                cons.append(review_cons[0].text)

        ## Change the pages to Next Page
        if not review_soup.findAll(attrs={'data-tn-element':['next-page']}):
            break
        else:
            review_url_all = review_soup.findAll(attrs={'data-tn-element':['next-page']})
            review_url = 'http://www.indeed.com/'+ str(review_url_all[0]['href'])
```

```python
cmp = {'Company' : company, 
           'Reviewer' : reviewer,
           'Date' : date,
           'Review Title' : title, 
           'Rating' : rating, 
           'Review content' : content,   
           'Pros' : pros, 
           'Cons' : cons}

df = pd.DataFrame.from_dict(cmp)
```
