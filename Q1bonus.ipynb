```python
from bs4 import BeautifulSoup
import urllib
import re
import pandas as pd
import requests
from urllib.request import urlopen
```

```python
#San Francisco
title_sf = []
company_sf = []
location_sf = []
jd_sf = []
url_sf = "https://www.indeed.com/m/jobs?q=data+scientist&l=san+francisco"
for i in range(20):
    
    page_sf = urlopen(url_sf)
    soup_sf = BeautifulSoup(page_sf, 'lxml')
    all_matches_sf = soup_sf.findAll(attrs={'rel':['nofollow']})
    for each in all_matches_sf:
        jd_url_sf= 'http://www.indeed.com/m/'+each['href']
        jd_page_sf =urlopen(jd_url_sf)
        jd_soup_sf = BeautifulSoup(jd_page_sf, 'lxml')
        jd_desc_sf = jd_soup_sf.findAll(attrs={'id':['desc']})
        title_sf.append(jd_soup_sf.body.p.b.font.text)
        company_sf.append(jd_desc_sf[0].span.text)
        location_sf.append(jd_soup_sf.body.p.span.text)
        jd_sf.append(jd_desc_sf[0].text)
        
    ## Change the pages to Next Page
    url_all_sf = soup_sf.findAll(attrs={'rel':['next']})
    url_sf = 'http://www.indeed.com/m/'+ str(url_all_sf[0]['href'])
```

```python
job_sf = {'Title': title_sf,
         'Company': company_sf,
         'Location': location_sf,
         'Job Description': jd_sf}
df_sf = pd.DataFrame.from_dict(job_sf)
```
