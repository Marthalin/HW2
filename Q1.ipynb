Question 1 :
Visit https://www.indeed.com/m/ website and search for Data Scientist in Los Angles. 
Now click on the “Find Jobs” button and you will be able to see the search results. 
Please find the top 200 jobs and obtain the frequencies of jobs that mention 
C, C++, Java, Javascript, Python, R, SQL, Hadoop, Hive, Pig, Spark, AWS and Tableau, respectively.

```python
from bs4 import BeautifulSoup
import urllib
import re
import pandas as pd
import requests
from urllib.request import urlopen
```

```python
title = []
company = []
location = []
jd = []
url = "https://www.indeed.com/m/jobs?q=data+scientist&l=Los+Angeles%2C+CA"
for i in range(20):
    
    page = urlopen(url)
    soup = BeautifulSoup(page, 'lxml')
    all_matches = soup.findAll(attrs={'rel':['nofollow']})
    for each in all_matches:
        jd_url= 'http://www.indeed.com/m/'+each['href']
        jd_page =urlopen(jd_url)
        jd_soup = BeautifulSoup(jd_page, 'lxml')
        jd_desc = jd_soup.findAll(attrs={'id':['desc']})
        title.append(jd_soup.body.p.b.font.text)
        company.append(jd_desc[0].span.text)
        location.append(jd_soup.body.p.span.text)
        jd.append(jd_desc[0].text)
        
    ## Change the pages to Next Page
    url_all = soup.findAll(attrs={'rel':['next']})
    url = 'http://www.indeed.com/m/'+ str(url_all[0]['href'])
```

```python
job = {'Title': title,
         'Company': company,
         'Location': location,
         'Job Description': jd}
df = pd.DataFrame.from_dict(job)
```

```python
description = pd.Series(jd)

mention_c = description.str.contains(r'\bC\b').sum()/200
mention_cplus = description.str.contains(r'\bC\+\+\b').sum()/200
mention_java = description.str.contains(r'\bJava\b').sum()/200
mention_javascript = description.str.contains(r'\bJavaScript\b').sum()/200
mention_python = description.str.contains(r'\bPython\b').sum()/200
mention_r = description.str.contains(r'\bR\b').sum()/200
mention_sql = description.str.contains(r'\bSQL\b').sum()/200
mention_hadoop = description.str.contains(r'\bHadoop\b').sum()/200
mention_hive = description.str.contains(r'\bHive\b').sum()/200
mention_pig = description.str.contains(r'\bPig\b').sum()/200
mention_spark = description.str.contains(r'\bSpark\b').sum()/200
mention_aws = (description.str.contains(r'\bAWS\b').sum()+description.str.contains(r'\bAmazon Web Services\b').sum())/200
mention_tableau = description.str.contains(r'\bTableau\b').sum()/200
```

```python
tool = ['C', 'C++', 'Java', 'Javascript', 'Python', 'R', 'SQL', 'Hadoop',
        'Hive', 'Pig', 'Spark', 'AWS', 'Tableau']
frequency = [mention_c, mention_cplus, mention_java, mention_javascript, 
            mention_python, mention_r, mention_sql, mention_hadoop, mention_hive, 
            mention_pig, mention_spark, mention_aws, mention_tableau]
table = {'Tool' : tool, 'Frequency' : frequency}

df = pd.DataFrame.from_dict(table)
df
```
